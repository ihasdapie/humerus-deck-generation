```
%tensorflow_version 1.x
!pip install -q gpt-2-simple
import gpt_2_simple as gpt2
from datetime import datetime
from google.colab import files
```

```
# fetch model
gpt2.download_gpt2(model_name="124M")
```

```
# grab question & answer datasets
!wget -q -O answers.txt https://drive.google.com/uc?id=1zff2jskQe244TiDV0Fr68khnRBrulfkM
!wget -q -O questions.txt https://drive.google.com/uc?id=1HYoV5SIymunvQDMy4-KNKFiIr_gBBrpw 
```

```
sess = gpt2.start_tf_sess()

gpt2.finetune(
    sess,
    dataset="answers.txt",
    steps=1000,
    restore_from="fresh",
    run_name="run2",
    print_every=10,
    sample_every=200,
    save_every=500,
)
```

```
gpt2.generate(sess, run_name='run2', top_k=50, temperature=0.8)
```
